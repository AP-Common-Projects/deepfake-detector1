```python
import json

file_path = "/content/drive/MyDrive/Colab Notebooks/medgemma_colab.ipynb"

def scrub_widgets(obj):
    if isinstance(obj, dict):
        if 'widgets' in obj:
            print(f"Found and removed 'widgets' key.")
            del obj['widgets']
        for key in list(obj.keys()):
            scrub_widgets(obj[key])
    elif isinstance(obj, list):
        for item in obj:
            scrub_widgets(item)

with open(file_path, 'r') as f:
    nb = json.load(f)

scrub_widgets(nb)

with open(file_path, 'w') as f:
    json.dump(nb, f, indent=1)

print("Deep scrub complete. Please Save (Ctrl+S) and check GitHub again.")
```
